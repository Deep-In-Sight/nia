```python
dtype = ["CamAngle", "DistDisp2Face", "DistCam2Face", "Eye-tracker", "FaceAngle"][0]
flist = glob(f"../nia/metadata/S1/002/T1/*/{dtype}/*.csv")

len(flist)
```

```python
ncol = 3
for ff in flist:
    #csv = pd.read_csv(ff)
    #arr = np.array(csv)[:,0]
    rrr = csv.reader(open(ff, "r"), quoting=csv.QUOTE_NONNUMERIC)
    all_lines = []
    for rr in rrr:
        all_lines.append(rr)
        
    arr = np.array(all_lines[1:]).astype(float) #####
    nn = len(arr)
    x = np.arange(nn)

    inter = interpolate.interp1d(x, arr[:,0])
    x_new = np.linspace(0, nn-1, 450)    
    c1 = inter(x_new)

    new_fn = ff.replace("S1/", "S1_new/")
    
    if ncol >= 2:
        inter = interpolate.interp1d(x, arr[:,1])
        c2 = inter(x_new)

    if dtype == "Eye-tracker":
        with open(new_fn, 'w') as f:
            f.write("[point] x   y\n")
            for v1,v2 in zip(c1,c2):
                f.write(f"{int(v1)}, {int(v2)}\n")
        
    if ncol == 3:
        inter = interpolate.interp1d(x, arr[:,2])
        c3 = inter(x_new)   
    
        with open(new_fn, 'w') as f:
            f.write("[CamAngle] Roll Pitch Yaw\n")
            for v1,v2,v3 in zip(c1,c2,c3):
                f.write(f"{v1:.3f}, {v2:.3f}, {v3:.3f}\n")
```
